### QUESTION 8

````python
result_8 = db.eu.find({ '$expr': {'$gt': [ {"$subtract":["$VALUE_EURO", "$AWARD_VALUE_EURO"]} , 10000000] }}).count()
result_8
````

In this question we are supposed to arrive at the total number of contracts were the difference between Proposed value (‘VALUE_EURO’) and awarded value (‘AWARD_VALUE_EURO’) is greater than 10000000. So we start by using a expression were we want the value inside the "gt" (greater than) expression , that corresponds to the subtraction of the "VALUES_EURO" with the "AWARD_VALUE_EURO" fields, to be greater than 10000000. Finally to get to our final answer we made the count operator.

### QUESTION 9

````python
result_9 = list(db.eu.aggregate([
    {"$group" : {'_id':"$ISO_COUNTRY_CODE", 'count':{'$sum':1}}}
    ,{ "$sort": {"count": -1 }},{"$limit": 1}
]))
result_9
````

To answer this and the following two questions, we make use of aggregate functions.
For question 9, we first group it by the value of ISO_COUNTRY_CODE, as this will represent a country’s designation, and by its count. On the latter part we are essentially saying each row represents a 1 and adding all those values. After this is done, we get an output which has the count of all countries’ contracts. To check which one has the greatest number of contracts we sort it by -1, which means that we sort it in descending, and limit the values to only 1 row. This outputs the country alongside the number of contracts.


### QUESTION 10

````python
result_10 = list(db.eu.aggregate([
    {"$group" : {'_id':"$YEAR", 'total':{'$max':"$AWARD_VALUE_EURO"}}}
    ,{ "$sort": {"total": -1 }},{"$limit": 1}
]))
result_10
````

For question 10 the process is quite like previous question. We this time group it by year and its maximum value. As we want the highest value per year, to then compare all other years, it only makes sense we use max, as this will output the max value per year of AWARD_VALUE_EURO. Then, like in the last question we sort it by the value of this new value outputted and limit it only to the highest one.

### QUESTION 11

````python
result_11 = list(db.eu.aggregate([
       {
     '$unwind': { 'path': "$AWARD_VALUE_EURO"}
   },
    { "$group": {"_id": {"year": "$YEAR",'country':"$ISO_COUNTRY_CODE","CPV": {'$substr':[{'$toString':'$CPV'},0,2]}},"avg" : {"$avg":"$AWARD_VALUE_EURO"}, "max" : {"$max":'$AWARD_VALUE_EURO'},"count": {"$sum": 1 }
    }},{ "$sort": {"count": -1 }}
]))
````

````python
result_11
````

Finally, in response to question 11, we first unwind AWARD_VALUE_EURO. This ensures that we will not be dealing with any rows that do not have that value, since on default unwind outputs all the rows that have the value present in path. After doing this we apply our criteria. We once again use a group to get the variables year, country and cpv code, assigning these 3 to be the id for the question, alongside the average and max of the contracts value, and the contracts count. To get the CPV code we use a substring function as we know that this code is present on the first two characters of the value, though we first have to convert it into a string, as the usage of the substring function is not possible with a number. After that we simply apply the avg and max functions to AWARD_VALUE_EURO and a count, which is explained in question 9. Finally, we just sort this by count so we can have some order to our list, although this last part is not necessary.
